```python
import csv
import numpy as np
```

```python
def fun1(file,col):
    with open(file, 'r', encoding='utf-8') as file:
        reader = csv.reader(file)
        data = np.array(list(reader))  
        header = data[0] 
        col_index = list(header).index(col)  

        col_data = data[1:, col_index]
        mask = col_data != ''  
        filtered_data = col_data[mask]
        
        res = np.array(filtered_data, dtype=float) 
        return res
fun1("russia_losses_equipment.csv","special equipment")
```

```python
zav2=fun1("russia_losses_equipment.csv","tank")
zav2
```

```python
def fun5(file):
    with open(file, 'r', encoding='utf-8') as file:
        reader = csv.reader(file)
        data = np.array(list(reader)) 
        date_data = data[1:, 0]
        start="2023-09-01"
        end="2023-11-30"
        start_index = list(date_data).index(start)
        end_index=list(date_data).index(end)

        header = data[0]
        tank_index = list(header).index("tank")
        tank_data = data[1:, tank_index]

        
        return int(tank_data[end_index])-int(tank_data[start_index])
zav5=fun5("russia_losses_equipment.csv")
zav5
```
